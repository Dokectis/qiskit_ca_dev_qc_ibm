# вҡӣпёҸ Quantum Entanglement: Creating the Bell State |ОҰ+вҹ©
## рҹҮӘрҹҮё Entrelazamiento CuГЎntico: Creando el Estado de Bell |ОҰ+вҹ©

In this notebook, we implement the most famous example of quantum entanglement: the **Bell State**. Entanglement is a unique quantum phenomenon where the status of two or more particles is linked, such that the state of one cannot be described independently of the other, even when separated by large distances.

**Core Concepts:**
1.  **Hadamard Gate (H):** Creates superposition.
2.  **Controlled-NOT Gate (CNOT):** Entangles two qubits.
3.  **Bell State |ОҰ+вҹ©:** A state where measuring one qubit immediately tells you the state of the other (100% correlation).

---

### рҹҮӘрҹҮё Conceptos Clave

En este notebook, implementamos el ejemplo mГЎs famoso de entrelazamiento cuГЎntico: el **Estado de Bell**. El entrelazamiento es un fenГіmeno cuГЎntico Гәnico donde el estado de dos o mГЎs partГӯculas estГЎ vinculado, de modo que el estado de una no puede describirse independientemente de la otra, incluso cuando estГЎn separadas por grandes distancias.

**Conceptos Principales:**
1.  **Puerta Hadamard (H):** Crea superposiciГіn.
2.  **Puerta NOT-Controlada (CNOT):** Entrelaza dos qubits.
3.  **Estado de Bell |ОҰ+вҹ©:** Un estado donde medir un qubit te indica inmediatamente el estado del otro (correlaciГіn del 100%).

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# [EN]: Initialize the Quantum Circuit with 2 qubits and 2 classical bits
# [ES]: Inicializar el Circuito CuГЎntico con 2 qubits y 2 bits clГЎsicos
qc = QuantumCircuit(2, 2)

print("Circuit initialized. / Circuito inicializado.")
```

```python
# [EN]: Step 1: Put Qubit 0 into superposition using the Hadamard gate
# [ES]: Paso 1: Poner el Qubit 0 en superposiciГіn usando la puerta Hadamard
qc.h(0)

# [EN]: Step 2: Entangle Qubit 0 with Qubit 1 using the CNOT gate
# [ES]: Paso 2: Entrelazar el Qubit 0 con el Qubit 1 usando la puerta CNOT
# (Control = Qubit 0, Target = Qubit 1)
qc.cx(0, 1)

# [EN]: Step 3: Add measurements
# [ES]: Paso 3: AГұadir mediciones
qc.measure([0, 1], [0, 1])

print("Circuit construction complete. / ConstrucciГіn del circuito completa.")
print(qc.draw(output='text'))
```

```python
# [EN]: Execute the simulation using AerSimulator
# [ES]: Ejecutar la simulaciГіn usando AerSimulator
simulator = AerSimulator()
job = simulator.run(qc, shots=1024)
result = job.result()
counts = result.get_counts(qc)

print(f"\nSimulation Counts / Conteos de SimulaciГіn: {counts}")

# [EN]: Visualize the distribution
# [ES]: Visualizar la distribuciГіn
plot_histogram(counts)
plt.show()
```
